```python
#function to remove asterisks
def remove_asterisks(s):
  if not isinstance(s, float) and '*' in s:
    return s.replace('*', "")
  return s

#function to convert object type to string type for mold species
def convert_to_float_or_zero(value):
  try:
    return float(value)
  except ValueError:
    return 0
for column_name in df.columns[3:41]:
    df[column_name] = df[column_name].apply(remove_asterisks)
    df[column_name] = df[column_name].apply(convert_to_float_or_zero)
    # df[column_name] = df[column_name].fillna(0)
df.head()
```

```python
#PCA
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaled_data = scaler.fit_transform(mold_species)

n_components = 2

pca = PCA(n_components=n_components)
pca_result = pca.fit_transform(scaled_data)

explained_variance = pca.explained_variance_ratio_
print("Explained Variance Ratios:", explained_variance)
print("Total Explained Variance:", sum(explained_variance))

import matplotlib.pyplot as plt

plt.scatter(pca_result[:, 0], pca_result[:, 1])
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('PCA Result')
plt.show()
```
